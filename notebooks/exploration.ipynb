```python
# Q2. Extract 2x2 patches, with a stride of 1
X = np.random.randint(0, 255, (8, 1, 32, 32))
```

```python
# Q3. We skip pixels here
image = np.random.randint(0, 256, (8, 1, 32, 32))
kernel_size, padding, stride = 2, 1, 2
#Â Pad the image
padded_image = np.pad(
    image,
    ((0, 0), (0, 0), (padding, padding), (padding, padding))
)
# Get the output shapes
B, C, H, W = X.shape
h_out = (H + 2*padding - kernel_size) // stride + 1
w_out = (W + 2*padding - kernel_size) // stride + 1
# Get the strides
s0, s1, s2, s3 = padded_image.strides
patches = np.lib.stride_tricks.as_strided(
    padded_image,
    strides = (s0, s2*stride, s3*stride, s2, s3, s1),
    shape = (B, h_out, w_out, kernel_size, kernel_size, C)
)
assert patches.shape == (8, 17, 17, 2, 2, 1)
print(patches.shape)
patches = np.reshape(patches, (B, h_out*w_out, kernel_size*kernel_size*C))
print(patches.shape)
```
